### Question 4

Join the data containing card details with transaction details so rows from the first which have matching ones in 
the second are returned.

Hint: `semi` is not an allowed option for `how` in `.merge()`. You'll need to think about the process of a semi 
join and then perform each step.

```python
# Using 'inner' as the how as this uses the intersection of the left & Right df's. left_on/right_on used as names are different.
cards.merge(transactions, how='inner', left_on='Card_Number', right_on='Credit_Card_ID')
```

### Question 5

Read in hat_observations and split `observation` into two columns, `hat_colour` and `hat_type`.

Hint: Look up the pandas documentation for `.str.split()`

```python
# import required data & view
hats = pd.read_csv("data/hat_observations.csv")
hats
```

```python
# Adding two new columns, colour & type, to the existing hats dataframe. Split at the comma.
hats[['hat_colour','hat_type']] = hats.observation.str.split(",", expand = True)

# Dropping old column
hats.drop(columns =["observation"], inplace = True)
 
# View expanded dataframe
hats
```
